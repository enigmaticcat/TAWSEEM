# Đếm số lượng giá trị OL và X trong các tệp CSV của PROVEDIt
Notebook này tính toán số lần xuất hiện của các giá trị `OL` và `X` trong tất cả các tệp CSV 1-5 person ở các mức số giây chạy khác nhau cho cả 5 thư mục thí nghiệm.

```python
import os
import pandas as pd

base_dir = '/Users/nguyenthithutam/Desktop/TAWSEEM/PROVEDIt_1-5-Person CSVs Filtered'
```

```python
def count_values(main_folder):
    main_path = os.path.join(base_dir, main_folder)
    if not os.path.isdir(main_path):
        print(f'{main_folder} not found.')
        return
    
    results = {}
    print(f'=== Kết quả cho thư mục: {main_folder} ===')
    for root, dirs, files in os.walk(main_path):
        for file in files:
            if file.endswith('.csv'):
                file_path = os.path.join(root, file)
                try:
                    df = pd.read_csv(file_path, low_memory=False)
                    # Tính chính xác số lần xuất hiện 'OL' và 'X'
                    ol_count = (df == 'OL').sum().sum()
                    x_count = (df == 'X').sum().sum()
                    
                    rel_path = os.path.relpath(root, main_path)
                    parts = rel_path.split(os.sep)
                    if len(parts) >= 2:
                        person_dir = parts[0]
                        sec_dir = parts[1]
                        if person_dir not in results:
                            results[person_dir] = {}
                        if sec_dir not in results[person_dir]:
                            results[person_dir][sec_dir] = {'OL': 0, 'X': 0, 'files': 0}
                        
                        results[person_dir][sec_dir]['OL'] += int(ol_count)
                        results[person_dir][sec_dir]['X'] += int(x_count)
                        results[person_dir][sec_dir]['files'] += 1
                except Exception as e:
                    pass
    
    # In kết quả đã sắp xếp
    for person in sorted(results.keys()):
        print(f'  {person}:')
        def get_sec(s):
            try: return int(s.split()[0])
            except: return 0
        for sec in sorted(results[person].keys(), key=get_sec):
            res = results[person][sec]
            print(f'    {sec}: OL = {res["OL"]}, X = {res["X"]} (từ {res["files"]} tệp CSV)')
    print('\n')
```
